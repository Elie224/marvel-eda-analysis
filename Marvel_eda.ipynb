```python
# 7 Donner la table de contingence des variables 'ALIGN' et 'ALIVE' (fonction 'crosstab' de pandas) puis effectuer un test d'indépendance du Khi-deux à un seuil de 5% (formuler les hypothèses, mener le test à l'aide de la fonction 'chi2_contingency' du module scipy stats puis conclure).
import pandas as pd
table = pd.crosstab(marvel["ALIGN"], marvel["ALIVE"])
table
```

```python
# test d'indépendance du Khi-deux à un seuil de 5%
from scipy.stats import chi2_contingency
chi2, p_value, dof, expected = chi2_contingency(table)
print("statistique du test : ", chi2)
print("p-value:", p_value)
```

```python
""" conclusion: 
H0: les variables ALIGN et ALIVE sont indépendantes
H1: les variables ALIGN et ALIVE ne sont pas independants
au seuil de 5% c'est 0.05
ici 
si p_value < 0.05 on rejette H0
p_value >= 0.05 on ne rejette pas H0
ici  p- value = 1.2782862450286972e-36 < 0.05
alors on rejette l'hypothese d'independence on conclut donc qu'il existe une relation statistiquement significative entre les deux variable
"""
```

```python
# 11 Créer une nouvelle variable 'Decade' qui indiquera la décennie de la première apparition des personnages (on utilisera la méthode 'cut' de pandas). On notera '30s', '40s', etc. les modalités de cette variable. Étudier ensuite cette variable individuellement puis conjointement avec les précédentes.
```

```python
bins = [1930, 1940,1950, 1960,1970,1980,1990,2000,2010,2020]
labels = ["30s", "40s","50s","60s","70s","80s","90s","200s","2010s"]
marvel["Decale"] = pd.cut(marvel["Year"], bins = bins, labels = labels)
marvel
```

```python
marvel["Decade"] = (marvel["Year"] // 10) * 10

decade_counts = marvel["Decade"].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(15,5))
decade_counts.plot(kind="bar", ax=ax)

ax.set_title("Nombre de personnages introduits par décennie")
ax.set_xlabel("Décennie")
ax.set_ylabel("Effectif")

plt.show()
```

```python
table_decade_align = pd.crosstab(marvel["Decale"], marvel["ALIGN"])
table_decade_align
```

```python
table_decade_align.plot(kind = "bar", figsize = (15,5))
plt.title("ALIGN selon la decenie d'apparution")
plt.xlabel("Decenie")
plt.ylabel("Effectif")
plt.show
```
